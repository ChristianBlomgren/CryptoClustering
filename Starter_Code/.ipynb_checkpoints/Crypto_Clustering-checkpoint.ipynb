```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
plot = df_market_data.hvplot.line(width=800, height=400, rot=90)
hvplot.save(plot, 'plot_image.png', fmt='png')
plot
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(df_market_data[["price_change_percentage_24h", 
                                                                    "price_change_percentage_7d", 
                                                                    "price_change_percentage_14d", 
                                                                    "price_change_percentage_30d", 
                                                                    "price_change_percentage_60d",
                                                                    "price_change_percentage_200d", 
                                                                    "price_change_percentage_1y"]])

# Make sure scaling was done correctly
market_data_scaled[0:5]
```

```python
# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(market_data_scaled, columns = ["price_change_percentage_24h", 
                                                                    "price_change_percentage_7d", 
                                                                    "price_change_percentage_14d", 
                                                                    "price_change_percentage_30d", 
                                                                    "price_change_percentage_60d",
                                                                    "price_change_percentage_200d", 
                                                                    "price_change_percentage_1y"])

# Copy the crypto names from the original data
df_market_data_scaled["coin_id"] = df_market_data.index
# Set the coinid column as index
df_market_data_scaled = df_market_data_scaled.set_index("coin_id")

# Display sample data
df_market_data_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
for k in range(1, 12):
    # Create a KMeans model using the loop counter for the n_clusters
    kmeans_model = KMeans(n_clusters=k, random_state=0)
    # Fit the model to the data using `df_market_data_scaled`
    kmeans_model.fit(df_market_data_scaled)
    # 3. Append the model.inertia_ to the inertia list
    inertia_values.append(kmeans_model.inertia_)
print(inertia_values)
```

```python
# Create a dictionary with the data to plot the Elbow curve
dict_elbow = {"k": list(range(1, 12)), "inertia": inertia_values}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(dict_elbow)
print(df_elbow)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
plt.figure(figsize=(10, 6))
plt.plot(df_elbow['k'], df_elbow['inertia'], marker='o', color='b', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.grid(True)
plt.show()
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** It appears that the best value for 'k' is at the 4 clusters point.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
best_k = 4
kmeans_model = KMeans(n_clusters=best_k, random_state=0)
```

```python
# Fit the K-Means model using the scaled data
kmeans_model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
cluster_labels = kmeans_model.predict(df_market_data_scaled)

# Print the resulting array of cluster values.
print(cluster_labels)
```

```python
# Create a copy of the DataFrame
df_market_data_scaled_predicitions = df_market_data_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_data_scaled_predicitions["cluster_labels"] = cluster_labels

# Display sample data
df_market_data_scaled_predicitions.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
scatter_plot = df_market_data_scaled_predicitions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="cluster_labels",
    hover_cols=["crypto_name"]
)

hvplot.save(scatter_plot, 'scatter_plot.png', fmt='png')
scatter_plot
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
pca_data = pca.fit_transform(df_market_data_scaled)
pca_df = pd.DataFrame(data=pca_data, columns=['PCA1', 'PCA2', 'PCA3'])
# View the first five rows of the DataFrame. 
print(pca_df.head())
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
explained_variance_ratio = pca.explained_variance_ratio_

print("Explained Variance Ratio for each Principal Component:")
for i, ratio in enumerate(explained_variance_ratio):
    print(f"PCA{i+1}: {ratio}")
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** ~0.895

```python
# Create a new DataFrame with the PCA data.

# Creating a DataFrame with the PCA data
pca_df_with_names = pd.DataFrame(data=pca_data, columns=['PCA1', 'PCA2', 'PCA3'])
# Copy the crypto names from the original data
pca_df_with_names['crypto_name'] = df_market_data_scaled.index

# Set the coinid column as index
pca_df_with_names.set_index('crypto_name', inplace=True)

# Display sample data
print(pca_df_with_names.head())
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k1 = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia_values1 = []

# Create a for loop to compute the inertia with each possible value of k
for k1 in range(1, 12):

    # 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans_model_pca = KMeans(n_clusters=k1, random_state=0)
    # 2. Fit the model to the data using `pca_df_with_names`
    kmeans_model_pca.fit(pca_df_with_names)
    # 3. Append the model.inertia_ to the inertia list
    inertia_values1.append(kmeans_model_pca.inertia_)
print(inertia_values1)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_pca = {"k": list(range(1, 12)), "inertia": inertia_values1}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow_pca = pd.DataFrame(elbow_data_pca)
print(df_elbow_pca)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
plt.figure(figsize=(10, 6))
plt.plot(df_elbow_pca['k'], df_elbow_pca['inertia'], marker='o', color='b', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Curve for PCA Data')
plt.grid(True)
plt.show()
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** It appears to be 4 clusters here as well.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
best_k_pca = 4
```

```python
# Fit the K-Means model using the PCA data
kmeans_model_pca.fit(pca_df_with_names)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
cluster_labels_pca = kmeans_model_pca.predict(pca_df_with_names)
# Print the resulting array of cluster values.
print(cluster_labels_pca)
```

```python
# Create a copy of the DataFrame with the PCA data
pca_df_with_names_copy = pca_df_with_names.copy()

# Add a new column to the DataFrame with the predicted clusters
pca_df_with_names_copy['predicted_cluster'] = cluster_labels_pca

# Display sample data
print(pca_df_with_names_copy.head())
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
scatter_plot_pca = pca_df_with_names_copy.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="predicted_cluster",
    hover_cols=["crypto_name"]
)

hvplot.save(scatter_plot_pca, 'scatter_plot_pca.png', fmt='png')
scatter_plot_pca
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves

# Plot the Elbow curve for the original data
plt.figure(figsize=(12, 6))
plt.plot(df_elbow['k'], df_elbow['inertia'], marker='o', color='b', linestyle='-', label='Original Data')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Curve Comparison')
plt.grid(True)

# Plot the Elbow curve for the PCA-transformed data
plt.plot(df_elbow_pca['k'], df_elbow_pca['inertia'], marker='s', color='r', linestyle='--', label='PCA Data')

plt.legend()
plt.show()
```

```python
# Composite plot to contrast the clusters
# Create subplots for the original data and PCA-transformed data clusters

# Plot the original data clusters
scatter_plot_original = df_market_data_scaled_predicitions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="cluster_labels",
    hover_cols=["crypto_name"]
)

# Plot the PCA-transformed data clusters
scatter_plot_pca = pca_df_with_names_copy.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="predicted_cluster",
    hover_cols=["crypto_name"]
)

# Overlay the two plots
composite_plot = scatter_plot_original * scatter_plot_pca

hvplot.save(composite_plot, 'composite_plot.png', fmt='png')
composite_plot
```
